```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение файла VCF
vcf_file = '/content/genotek.vcf'
data = []

with open(vcf_file, 'r') as file:
    for line in file:
        if not line.startswith('#'):  # Пропускаем строки с метаданными
            parts = line.strip().split('\t')
            data.append(parts)

# Преобразование данных в DataFrame
columns = ['#CHROM', 'POS', 'ID', 'REF', 'ALT', 'QUAL', 'FILTER', 'INFO', 'FORMAT', 'dd5984']
df = pd.DataFrame(data, columns=columns)

# Функция для определения типа генотипа
def determine_genotype(genotype):
    try:
        alleles = genotype.split('/')
        if alleles[0] == alleles[1]:
            return 'homozygous'
        else:
            return 'heterozygous'
    except:
        return 'unknown'

# Применение функции к столбцу генотипов
df['Genotype_Type'] = df['dd5984'].apply(determine_genotype)

# Группировка по хромосомам и типу генотипа
genotype_counts = df.groupby(['#CHROM', 'Genotype_Type']).size().unstack(fill_value=0)

# Рассчитываем общее количество генотипов для каждой хромосомы
total_genotypes = genotype_counts.sum(axis=1)

# Рассчитываем соотношение гомозигот и гетерозигот
genotype_ratios = genotype_counts.div(total_genotypes, axis=0)

# Добавляем процентное соотношение гетерозигот
genotype_ratios['Heterozygosity_Percent'] = genotype_ratios['heterozygous'] * 100

# Рассчитываем среднюю гетерозиготность по всем хромосомам
mean_heterozygosity = genotype_ratios['heterozygous'].mean() * 100

# Определяем порядок хромосом
chrom_order = ['chr' + str(i) for i in range(1, 23)] + ['chrX', 'chrY', 'chrM']

# Сортируем DataFrame по определенному порядку хромосом
genotype_ratios = genotype_ratios.reindex(chrom_order)

# Создаем график
plt.figure(figsize=(14, 8))

# Построение столбчатой диаграммы
bars = plt.bar(genotype_ratios.index, genotype_ratios['Heterozygosity_Percent'], color='skyblue')

# Добавляем подписи к столбцам с процентным соотношением гетерозигот
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}%', ha='center', va='bottom')

# Устанавливаем подписи осей и заголовок
plt.xlabel('Хромосома')
plt.ylabel('% Гетерозигот')
plt.title('% Гетерозигот по хромосомам')

# Устанавливаем лимиты оси Y для лучшей читаемости
plt.ylim(0, max(genotype_ratios['Heterozygosity_Percent']) * 1.1)

# Добавляем сетку для лучшей ориентации
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Улучшаем отображение меток хромосом
plt.xticks(rotation=45)
plt.tight_layout()

# Сохранение графика
plt.savefig('genotype_heterozygosity.png')
plt.show()

# Вывод таблицы с соотношениями
print("Средняя гетерозиготность по всем хромосомам: {:.2f}%".format(mean_heterozygosity))
print(genotype_ratios[['homozygous', 'heterozygous', 'Heterozygosity_Percent']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Чтение файла VCF
vcf_file = '/content/snps_clean.vcf'
data = []

with open(vcf_file, 'r') as file:
    for line in file:
        if not line.startswith('#'):  # Пропускаем строки с метаданными
            parts = line.strip().split('\t')
            data.append(parts)

# Преобразование данных в DataFrame
columns = ['#CHROM', 'POS', 'ID', 'REF', 'ALT', 'QUAL', 'FILTER', 'INFO', 'FORMAT', 'dd5984']
df = pd.DataFrame(data, columns=columns)

# Удаляем строки, содержащие './.' в столбце dd5984
df = df[df['dd5984'] != './.']

# Функция для определения типа генотипа
def determine_genotype(genotype):
    if genotype == '--':
        return 'unknown'
    alleles = genotype.split('/')
    if len(alleles) != 2:
        return 'unknown'
    if alleles[0] == alleles[1]:
        return 'homozygous'
    else:
        return 'heterozygous'

# Применение функции к столбцу генотипов
df['Genotype_Type'] = df['dd5984'].apply(determine_genotype)

# Группировка по хромосомам и типу генотипа
genotype_counts = df.groupby(['#CHROM', 'Genotype_Type']).size().unstack(fill_value=0)

# Рассчитываем общее количество генотипов для каждой хромосомы
total_genotypes = genotype_counts.sum(axis=1)

# Рассчитываем соотношение гомозигот и гетерозигот
genotype_ratios = genotype_counts.div(total_genotypes, axis=0).fillna(0)

# Добавляем процентное соотношение гетерозигот
genotype_ratios['Heterozygosity_Percent'] = genotype_ratios['heterozygous'] * 100

# Определяем порядок хромосом
chrom_order = ['chr' + str(i) for i in range(1, 23)] + ['chrX', 'chrY', 'chrM']

# Переименовываем хромосомы в DataFrame для соответствия порядку
genotype_ratios.index = 'chr' + genotype_ratios.index.astype(str)

# Сортируем DataFrame по определенному порядку хромосом
genotype_ratios = genotype_ratios.reindex(chrom_order).fillna(0)

# Создаем график
plt.figure(figsize=(14, 8))

# Построение столбчатой диаграммы
bars = plt.bar(genotype_ratios.index, genotype_ratios['Heterozygosity_Percent'], color='skyblue')

# Добавляем подписи к столбцам с процентным соотношением гетерозигот
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}%', ha='center', va='bottom')

# Устанавливаем подписи осей и заголовок
plt.xlabel('Хромосома')
plt.ylabel('% Гетерозигот')
plt.title('% Гетерозигот по хромосомам')

# Устанавливаем лимиты оси Y для лучшей читаемости
max_heterozygosity = genotype_ratios['Heterozygosity_Percent'].max()
if pd.isna(max_heterozygosity) or not np.isfinite(max_heterozygosity):
    max_heterozygosity = 100  # Устанавливаем максимальное значение в 100%, если оно NaN или Inf
plt.ylim(0, max_heterozygosity * 1.1)

# Добавляем сетку для лучшей ориентации
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Улучшаем отображение меток хромосом
plt.xticks(rotation=45)
plt.tight_layout()

# Сохранение графика
plt.savefig('genotype_heterozygosity.png')
plt.show()

# Вывод таблицы с соотношениями
print("Средняя гетерозиготность по всем хромосомам: {:.2f}%".format(genotype_ratios['Heterozygosity_Percent'].mean()))
print(genotype_ratios[['homozygous', 'heterozygous', 'Heterozygosity_Percent']])
```

```python
import pandas as pd

# Чтение первого файла VCF
vcf_file1 = '/content/genotek.vcf'
data1 = []

with open(vcf_file1, 'r') as file:
    for line in file:
        if not line.startswith('#'):  # Пропускаем строки с метаданными
            parts = line.strip().split('\t')
            data1.append(parts)

# Преобразование данных первого файла в DataFrame
columns1 = ['#CHROM', 'POS', 'ID', 'REF', 'ALT', 'QUAL', 'FILTER', 'INFO', 'FORMAT', 'dd5984']
df1 = pd.DataFrame(data1, columns=columns1)

# Чтение второго файла VCF
vcf_file2 = '/content/snps_clean.vcf'
data2 = []

with open(vcf_file2, 'r') as file:
    for line in file:
        if not line.startswith('#'):  # Пропускаем строки с метаданными
            parts = line.strip().split('\t')
            data2.append(parts)

# Преобразование данных второго файла в DataFrame
columns2 = ['#CHROM', 'POS', 'ID', 'REF', 'ALT', 'QUAL', 'FILTER', 'INFO', 'FORMAT', 'dd5984']
df2 = pd.DataFrame(data2, columns=columns2)

# Преобразование названий хромосом во втором файле к формату первого файла
df2['#CHROM'] = 'chr' + df2['#CHROM'].astype(str).replace({'chrMT': 'chrM'})

# Подсчет количества строк в каждом файле
count_rows_file1 = len(df1)
count_rows_file2 = len(df2)

# Находим записи с одинаковыми SNP и одинаковыми аллелями
common_snps_same_alleles = pd.merge(df1, df2, on=['#CHROM', 'ID', 'dd5984'], how='inner')

# Находим уникальные записи в каждом файле
unique_snps_file1 = df1[~df1.set_index(['#CHROM', 'ID', 'dd5984']).index.isin(df2.set_index(['#CHROM', 'ID', 'dd5984']).index)]
unique_snps_file2 = df2[~df2.set_index(['#CHROM', 'ID', 'dd5984']).index.isin(df1.set_index(['#CHROM', 'ID', 'dd5984']).index)]

# Находим суммарное количество уникальных SNP в обоих файлах
unique_snps_combined = pd.concat([df1, df2]).drop_duplicates(subset=['#CHROM', 'ID'])

# Находим записи с одинаковыми SNP, но различными аллелями
common_snps_diff_alleles = pd.merge(df1, df2, on=['#CHROM', 'ID'], how='inner')
common_snps_diff_alleles = common_snps_diff_alleles[common_snps_diff_alleles['dd5984_x'] != common_snps_diff_alleles['dd5984_y']]

# Вывод результатов
print(f"Количество строк в первом файле: {count_rows_file1}")
print(f"Количество строк во втором файле: {count_rows_file2}")
print(f"Количество записей с одинаковыми SNP и одинаковыми аллелями: {len(common_snps_same_alleles)}")
print(f"Количество уникальных записей в первом файле: {len(unique_snps_file1)}")
print(f"Количество уникальных записей во втором файле: {len(unique_snps_file2)}")
print(f"Суммарное количество уникальных SNP в обоих файлах: {len(unique_snps_combined)}")
print(f"Количество записей с одинаковыми SNP, но различными аллелями: {len(common_snps_diff_alleles)}")

# Вывод примера записей с одинаковыми SNP и одинаковыми аллелями
print("\nПример записей с одинаковыми SNP и одинаковыми аллелями:")
print(common_snps_same_alleles.head())

# Вывод примера уникальных записей в первом файле
print("\nПример уникальных записей в первом файле:")
print(unique_snps_file1.head())

# Вывод примера уникальных записей во втором файле
print("\nПример уникальных записей во втором файле:")
print(unique_snps_file2.head())

# Вывод примера записей с одинаковыми SNP, но различными аллелями
print("\nПример записей с одинаковыми SNP, но различными аллелями:")
print(common_snps_diff_alleles.head())
```

```python
with open('genotek_eff.vcf', 'r') as file:
    ann_count = 0
    for line in file:
        if not line.startswith('#'):
            # Разделяем строку по табуляции
            fields = line.strip().split('\t')
            # Проверяем, есть ли поле ANN в информации о вариантах
            if len(fields) > 7 and 'ANN=' in fields[7]:
                ann_count += 1

print(f"Количество строк с ANN: {ann_count}")
```

```python
with open('genotek_eff.vcf', 'r') as file:
    HIGH_count = 0
    for line in file:
        if not line.startswith('#'):
            # Разделяем строку по табуляции
            fields = line.strip().split('\t')
            # Проверяем, есть ли HIGH в информации о вариантах
            if len(fields) > 7 and 'HIGH' in fields[7]:
                HIGH_count += 1

print(f"Количество строк с HIGH: {HIGH_count}")
```

```python
# Открываем исходный файл VCF и новый файл для записи
with open('genotek_eff.vcf', 'r') as infile, open('high_impact.vcf', 'w') as outfile:
    for line in infile:
        # Записываем заголовочные строки в новый файл без изменений
        if line.startswith('#'):
            outfile.write(line)
        else:
            # Разделяем строку по табуляции
            fields = line.strip().split('\t')
            # Проверяем, есть ли HIGH в информации о вариантах
            if len(fields) > 7 and 'HIGH' in fields[7]:
                outfile.write(line)

print("Строки с HIGH сохранены в файл high_impact.vcf")
```
